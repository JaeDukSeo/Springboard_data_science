## Data wrangling
1. Transform the data frame into a more workable state, usig Pivot_table() 
2. Interpolate missing values for the LifeExpectancy data set

```python
df_merged = pd.merge(df, metrics, on='IndicatorName')
df_merged
```

Just life expectancy data

```python
lexp = df_merged[(df_merged.IndicatorCode =='SP.DYN.LE00.IN')]
#lexp = lexp[(lexp.CountryName == 'World') | (lexp.CountryName == 'United States') | (lexp.CountryName == 'Haiti')]
lexp = lexp[lexp.CountryName.str.contains('income')]
lexp = lexp.pivot_table(values='Value', index=['Year'], columns=['CountryName'])
lexp.head()
```

```python
lexp.plot(figsize=(18,9))
```
